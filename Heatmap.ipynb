```python
only_state=city_state.groupby('state_l')['state_l'].count()
```

```python
only_state
```

```python
only_state_df=pd.DataFrame({
    'State':only_state.index,
    'Weight': only_state.values
})
```

```python
only_state_df.head()
```

```python
state_lat=[]
state_lng=[]
state_counter=0
```

```python
for states in only_state_df['State']:
    
    geo_data = requests.get(('https://maps.googleapis.com/maps/api/geocode/json?'
    'address={0}&key={1}').format(states, api_key)).json()
    #include print log for lat data
    print(f'Finding record:{state_counter} for {states} at lat:{geo_data["results"][0]["geometry"]["location"]["lat"]} and lng:{geo_data["results"][0]["geometry"]["location"]["lng"]}')
    state_lat.append(geo_data["results"][0]["geometry"]["location"]["lat"])
    state_lng.append(geo_data["results"][0]["geometry"]["location"]["lng"])
    state_counter=state_counter+1
```

```python
only_state_df['Lat']=state_lat
```

```python
only_state_df['Lng']=state_lng
```

```python
only_state_df.head()
```

```python
only_state_df.to_csv('Weighted_State_Lat_Lng.csv')
```

```python
locations = only_state_df[["Lat", "Lng"]].astype(float)
companies_per_state=only_state_df['Weight'].astype(float)
```

```python
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=companies_per_state, 
                                 dissipating=False)
heat_layer.dissipating = False
heat_layer.max_intensity = 4000
heat_layer.point_radius = 3

fig.add_layer(heat_layer)

fig
fig.
```
